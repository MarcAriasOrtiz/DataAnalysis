# Creación Dataset intervalo 1

```python
# · Utilizaremos pandas para crear el Dataset
import pandas as pd

# · Definimos los nuevos arrays que contendrán los datos del intervalo 1 de los pacientes
intervalo1BC = []
intervalo1mBC = []
```

```python
# · Definimos el número de iteraciones en base al tamaño de los datos del archivo .h5
iter1BC = (len(dataBC)) 
iter1mBC = (len(datamBC)) 

# · Hacemos un bucle para extraer solo los datos del intervalo 1 para BC y mBC
c = 0
while c != iter1BC:
    intervalo1BC.append(dataBC[c])
    c = c + 3

c = 0
while c != iter1mBC:
    intervalo1mBC.append(datamBC[c])
    c= c + 3
```

```python
# · Comprobamos que tenemos todos los pacientes debería ser igual a las comprobaciones anteriores hechas divididas entre 3
print('Pacientes BC del primer intervalo:')
print(len(intervalo1BC))
print('Pacientes mBC del primer intervalo:')
print(len(intervalo1mBC))
```

```python
# · Creamos el primer DataFrame con Pandas para BC del intervalo 1
PrimerIntBC = pd.DataFrame(intervalo1BC)
PrimerIntBC;
```

```python
# · Eliminamos la última columna para que todo el DataFrame tenga la misma dimensión
PrimerIntBC = PrimerIntBC.drop(columns=[4416])
PrimerIntBC;
```

```python
# · Añadimos la variable Target que determina si tiene enfermedad metastasica, 0 en caso de que sea BC y 1 en caso de mBC
Target = []

for i in range(len(intervalo1BC)):
    Target.append(0)

PrimerIntBC.loc[:,'Target'] = Target
PrimerIntBC;
```

```python
# · Creamos el DataFrame con Pandas para mBC del intervalo 1
PrimerIntmBC = pd.DataFrame(intervalo1mBC)
PrimerIntmBC;
```

```python
# · Eliminamos la última columna para que todo el DataFrame tenga la misma dimensión
PrimerIntmBC = PrimerIntmBC.drop(columns=[4416])
PrimerIntmBC;
```

```python
# · Añadimos la variable Target que determina si tiene enfermedad metastasica, 0 en caso de que sea BC y 1 en caso de mBC

Target = []

for i in range(len(intervalo1mBC)):
    Target.append(1)

PrimerIntmBC.loc[:,'Target'] = Target
PrimerIntmBC;
```

```python
# · Finalmente unímos en un único DataFrame con información de BC y mBC con su Target correspondiente
DatasetIntervalo1 = pd.concat([PrimerIntBC, PrimerIntmBC])
DatasetIntervalo1
```

# Comprobación de intervalos para la clasificación

Aplicaremos un Random Forest Regressor (RFR) sencillo para ver si tiene sentido utilizar todos los valores de la respresentación del espectro 1D en la resonancia magnética nuclear.  

# Aplicación de RFR  Intervalo 1

```python
from sklearn.model_selection import train_test_split

# · Definimos X e Y
X1 = DatasetIntervalo1[:4417];
X1 = X1.drop(columns=['Target']) # · Dejamos el dataset sin la variable de 'Target'

Y1 = DatasetIntervalo1.iloc[:,-1] # · Seleccionamos la columna 'Target'

# · Diividimos en Train test Split
X_train, X_test, y_train, y_test = train_test_split(
    X1, 
    Y1, 
    test_size=0.15, 
    random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor

# · Unos 19 - 20 min Videollamada en Zoom mientras corría.
rf1 = RandomForestRegressor(n_estimators = 10000, random_state = 42)
rf1.fit(X_train, y_train)
```

```python
predictions = rf1.predict(X_test)
Predictions = np.round(predictions)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, Predictions)
```

```python
# · Hacemos una matriz de confusión que sea más fácil de entender
data = {'y_Actual':    y_test,
        'y_Predicted': Predictions
        }

df = pd.DataFrame(data, columns=['y_Actual','y_Predicted'])

confusion_matrix1 = pd.crosstab(df['y_Actual'], df['y_Predicted'], rownames=['Actual'], colnames=['Predicted'])
print (confusion_matrix1)
```

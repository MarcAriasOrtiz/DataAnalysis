# Ejemplo de ajuste

**Veamos un ejemplo de ajuste de una funci√≥n a un conjunto de datos, que generaremos en primer lugar.**

```r
x<-seq(0,50,1)
y<-((runif(1,10,20)*x)/(runif(1,0,10)+x))+rnorm(51,0,1)
plot(x,y)
```

**La funci√≥n que ajustaremos es $y(x)=\frac{ax}{b+x}$, donde ponemos un valor inicial a los par√°metros $a=10$ y $b=1$. Para realizar el ajuste se utiliza el m√©todo de m√≠nimos cuadrados (nls).**

```r
m<-nls(y~a*x/(b+x), start=list(a=10,b=1))
summary(m)
```

**Una manera de comprobar la bondad del ajuste, es mirar la correlaci√≥n entre los valores predichos por el ajuste y los valores reales de y:**

```r
cor(y,predict(m))
```

**Y gr√°ficamente:**

```r
plot(x,y)
lines(x,predict(m),lty=2,col="red",lwd=3)
```

**Calculamos los residuales (diferencia entre curva ajustada y puntos, normalizado al error):**

```r
res = resid(m)
```

```r
plot(x, res, ylab="Residuales", xlab="x") 
abline(0, 0) 
```

**Como vemos, los valores est√°n entre -3 y 3, por lo que se considera un buen ajuste en todo el rango de x.**

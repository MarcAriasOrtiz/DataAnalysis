# Ejemplo de ajuste de una función

**Generaremos unos sucesos según una distribución normal y a continuación probaremos a ajustar diferentes funciones. Esta vez usando fitdist, que utiliza el método de maximum likelihood para el ajuste.**

```r
x <- rnorm(1000)
hist(x, prob=TRUE, freq=FALSE, col=rgb(1,0,0,0.5), breaks=seq(-5,5,by=0.5), xlim=c(-5,5))
```

```r
library(MASS)
require(fitdistrplus)
```

```r
f1 <- fitdist(x,"norm")#Normal
f2 <- fitdist(x,"logis")#Logística, parecida a la normal, pero de colas más pesadas
f3 <- fitdist(x,"cauchy")#Breit-Wigner
```

```r
summary(f1)
```

```r
summary(f2)
```

```r
summary(f3)
```

**Un método para visualizar el ajuste y obtener los Q-Q y P-P plots es usar plotdist:**

```r
plotdist(x,"norm",para=list(mean=f1$estimate[1],sd=f1$estimate[2]))
```

```r
plotdist(x,"logis",para=list(location=f2$estimate[1],scale=f2$estimate[2]))
```

```r
plotdist(x,"cauchy",para=list(location=f3$estimate[1],scale=f3$estimate[2]))
```

**Podemos ver que el mejor ajuste es el de la normal, como era de esperar, y que el resultado de usar la función logística es parecido, salvo en las colas.**

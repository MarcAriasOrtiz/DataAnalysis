Podemos observar en los datos que tenemos un total de 32 datos, cada dato tiene asignado un sexo, edad, si sobrevivió o no y la frecuencia. Este último parámetro indica el número de personas con esos parámetros que había en el Titanic. De esta manera, hay que ajustar el dataset para tener la cantidad de datos dependiendo de las frecuencias, asi poder realizar el algoritmo de la mejor manera y tener los resultados más fiables respecto a la realidad.

### Creación del DATASET 

```r
#Ahora repetiremos cada combinación igual a la frecuencia de cada fila
Frecuencia=rep.int(seq_len(nrow(Titanic)), Titanic$Freq) 
#Creamos el nuevo dataset teniendo encuenta la frecuencia
Titanic_dataset=Titanic[Frecuencia,]
#Commo ya hemos tenido en cuenta la frecuencia la podemos eliminar del dataset
Titanic_dataset$Freq=NULL
Titanic_dataset
summary(Titanic_dataset)
```

### División del DATASET - Train y Test

```r
set.seed(2021)
tamano.total=nrow(Titanic_dataset)
tamano.train=round(tamano.total*0.75)

Titanic_dataset.indices = sample(1:tamano.total , size=tamano.train)
Titanic_dataset.train = Titanic_dataset[Titanic_dataset.indices,]
Titanic_dataset.test = Titanic_dataset[-Titanic_dataset.indices,]
print("Este es el TRAIN")
summary(Titanic_dataset.train)
print("Este es el TEST")
summary(Titanic_dataset.test)
```

###  Clasificador Naive Bayes para la variable objetivo "Survived"

```r
modelo = naiveBayes(Survived ~ ., data=Titanic_dataset.train) 
modelo
```

### Predicción

```r
results = predict(object = modelo, newdata=Titanic_dataset.test, type = "class")
results
```

### Visualización de resultados

```r
tab = table(Titanic_dataset.test$Survived,results,dnn = c('REAL','PREDICCIÓN'))
confusionMatrix(tab) 
```

# Análisis y Conclusiones

A primera vista nos indica que el modelo tiene una "Accuracy" de 0,7954 que sería un 79,45% no parece del todo malo.
Si nos fijamos en los casos de "No Survived" apreciamos que de los 385 casos totales clasifica bien 355 y 30 mal. Esto supone una "Accuracy" del 92,2% (Muy buena). Por otra parte, si nos centramos en la predicción de los "Yes Survived" la "Accuracy" desciende mucho. Apreciamos que hay 165 casos que "Yes Survived" de los cuales, clasifica bien 82 y clasifica mal 83. Esto supone una "Accuracy" de 49,7% (Bastante mala).

Al observar las tablas hechas y el clasificador Naive-Bayes podemos apreciar como si eres niño, mujer o viajas en primera clase tienes más probabilidades de sobrevivir.

Naive Bayes es un algoritmo paramétrico, esto implica que, aunque hagamos muchas repeticiones, los resultados seguiran siendo los mismos. La única manera de mejorar los resultados o que sean diferentes, sería tener más parámetros a analizar o más cantidad de datos. También hay que tener en cuenta que los parámetros que analizamos para hacer el modelo son independientes entre si. A primera vista debería ser así, pero diría que el viajar a en primera clase o no parámetro muy decisivo en si sobrevives o no depende bastante de la edad. Ya que, proporcionalmente si eres adulto es más probable que viajes en primera clase.

95% CI : (0.7583, 0.8276) Intervalo de confianza Que nos muestra que la "Accuracy" de este modelo va del 75,8% al 82,8%

No Information Rate : 0.7964 esta es la precisión que se puede lograr al predecir siempre la etiqueta de la clase mayoritaria. En este caso, si se le pide que prediga si una persona sobrevive o no, al elegir siempre "No Survived", podemos lograr casi un 80% de precisión en el conjunto de prueba. Nuestra "Accuracy" no se ve tan bien ahora.

P-Value [Acc > NIR] : 0.5671  este p-value muestra que el modelo no ofrece un rendimiento significativo  sobre la tasa de ausencia de información.

Kappa : 0.4614 el estadístico de Kappa muestra qué tan bien nuestras predicciones de clasificadores coincidieron con las etiquetas de clase reales mientras se controla la precisión de un clasificador aleatorio. El valor de 0,46 representa una moderada coincidencia. Contra más cerca de 1 este este valor mejor.

Mcnemar's Test P-Value : 9.994e-07 Nos muestra la probabilidad de estar equivocados si concluimos que los parámetros que utilizamos para hacer el modelo para calcular la supervivencia están asociados al parámetro "Survived"

Sensitivity : 0.8105 se calcula como el número de predicciones "No survived" correctas dividido por el número total de "No Survived". Contra más cerca de 1 se encuentre mejor más sensibilidad tendrá nuestro modelo.

Specificity : 0.7321 se calcula como el número de predicciones "Yes Survived" correctas dividido por el número total de "Yes Survived". La mejor especifidad es un valor de 1. 

Pos Pred Value : 0.9221 se calcula como el número de predicciones positivas correctas dividido por el número total de predicciones positivas. También se le llama valor predictivo positivo. En este caso el PPV está calculado respecto al "No Survivied"

Neg Pred Value : 0.4970 se calcula como el número de predicciones positivas incorrectas dividido por el número total de negativas. En este caso el NPV está calculado respecto al "Yes Survived"

Prevalence : 0.7964 muestra la frecuencia en que aparecen los "No Survived" en este modelo.  

Detection Rate : 0.6455 muestra el número de predicciones de "No Survived" correctas hechas como una proporción de todas las predicciones hechas

Detection Prevalence : 0.7000 muestra el número de predicciones de clase "No Survived" hechas como proporción de todas las predicciones

Balanced Accuracy : 0.7713  esencialmente toma el promedio de las tasas verdaderas positivas y verdaderas negativas
